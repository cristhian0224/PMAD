## 1. Análisis Nacional de los tamaños de deforestación de la serie de tiempo 2013-2019

El archivo contiene las áreas (hectáreas) de cada polígono detectado como deforestación para cada periodo. Las columnas presentes en el archivo de la serie temporal son las siguientes:

+ Deforestación 2012 - 2013   =>   def_13
+ Deforestación 2013 - 2014   =>   def_14
+ Deforestación 2014 - 2015   =>   def_15

+ Deforestación 2017 - 2018   =>   def_18
